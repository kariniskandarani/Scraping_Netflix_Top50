```python
#The below code created for testing

#scraping website
# to obtain
# Title, Year, Poster, Duration, IMDB Rating, Genre, Stars (Actors), Director TEXT
import re
movies_data= []
for div_movie in div_movies:
    #obtain the title and year of the movie
    title = div_movie.find('h3', class_='b-media__title').get_text(strip=True)
    year = title.split()[-1]
    title = ' '.join(title.split()[:-1]) #removing the paranthesis & the year from the title

    #obtain the attribute (data-dt-lazy-src) value of the img
    poster = div_movie.find('img', class_='dt-lazy-load')['data-dt-lazy-src']

    # Find all divs with the class 'b-media__score'to obtain IMDB rate
    media_scores = div_movie.find_all('div', class_='b-media__score')
    if len(media_scores) >= 2:
      #get the IMDB rating
      imdb_rating = media_scores[1].get_text(strip=True)
    else:
      #imdb rate not found so none
      imdb_rating = None


    # Find the div with the class 'b-media__basic' to obtain duration
    duration_div = div_movie.find('div', class_='b-media__basic')
    if duration_div:
       # Remove the text of the 'b-media__rating' child div, if it exists
       rating_div = duration_div.find('div', class_='b-media__rating')
       if rating_div:
         rating_div.extract()  # Removes the child div from the parent (duration_div)
       #obtain the text content from the duration_div
       duration_txt = duration_div.get_text(strip=True)
       # Use regex to extract the numeric part
       match = re.search(r'\d+', duration_txt)
       if match:
        duration = int(match.group())
       else:
        duration = None

   #find all divs with class b-media__value
    media_div = div_movie.find_all('div', class_='b-media__value')
    #Obtain genre, stars, director
    genre = media_div[0].get_text(strip=True)
    stars = media_div[1].get_text(strip=True)
    director = media_div[2].get_text(strip=True)

    movies_data.append({
        'id': len(movies_data) + 1,
        'title': title,
        'year': year[1:-1],#to remove trailing paranthesis
       'poster': poster,
       'duration': duration,
       'imdb_rating': imdb_rating,
       'genre': genre,
       'stars': stars,
       'director': director
    })
```

```python


import re

# Function to extract individual movie details
def extract_movie_details(div_movie):
    """
    Extracts details (title, year, poster, duration, IMDB rating, genre, stars, director) for a single movie.
    """
    # Initialize variables
    title = year = poster = duration = imdb_rating = genre = stars = director = None

    # 1. Extract Title and Year
    title_div = div_movie.find('h3', class_='b-media__title')
    if title_div:
        title_text = title_div.get_text(strip=True)
        year = title_text.split()[-1] if title_text.split()[-1].startswith('(') else None
        title = ' '.join(title_text.split()[:-1])  # Remove the parentheses & year
    print(f"Title: {title}, Year: {year}")

    # 2. Extract Poster URL
    img_tag = div_movie.find('img', class_='dt-lazy-load')
    if img_tag and 'data-dt-lazy-src' in img_tag.attrs:
        poster = img_tag['data-dt-lazy-src']
    print(f"Poster URL: {poster}")

    # 3. Extract IMDB Rating
    media_scores = div_movie.find_all('div', class_='b-media__score')
    if len(media_scores) >= 2:
        imdb_rating_txt = media_scores[1].get_text(strip=True)
        # Use regex to extract the numeric part before "/10"
        match = re.match(r"(\d+(\.\d+)?)/10", imdb_rating_txt)
        if match:
            imdb_rating = float(match.group(1))
        else:
           imdb_rating = None
    print(f"IMDB Rating: {imdb_rating}")

    # 4. Extract Duration
    duration_div = div_movie.find('div', class_='b-media__basic')
    if duration_div:
        rating_div = duration_div.find('div', class_='b-media__rating')
        if rating_div:
            rating_div.extract()  # Remove the rating child div from the parent (duration_div)
        duration_txt = duration_div.get_text(strip=True)
        match = re.search(r'\d+', duration_txt)
        if match:
            duration = int(match.group())
    print(f"Duration: {duration}")

    # 5. Extract Genre, Stars, and Director
    #find all divs with class b-media__value
    media_divs = div_movie.find_all('div', class_='b-media__value')
    genre = media_divs[0].get_text(strip=True)
    stars = media_divs[1].get_text(strip=True)
    director = media_divs[2].get_text(strip=True)
    print(f"Genre: {genre}, Stars: {stars}, Director: {director}")

    # Return the extracted data
    return {
        'title': title,
        'year': year[1:-1] if year else None,  # Remove trailing parentheses
        'poster': poster,
        'duration': duration,
        'imdb_rating': imdb_rating,
        'genre': genre,
        'stars': stars,
        'director': director
    }


```

```python

# Function to process all movies and insert them to a database
def process_save_movies(div_movies):
    """
    Processes a list of div_movies, extracts details, and inserts them to a database.
    """
    movies_data = []

    # Processing each movie div
    for div_movie in div_movies:
        movie_details = extract_movie_details(div_movie)
        movies_data.append(movie_details)

    # Simulate saving to a database
    print("Saving the following data to the database:")
    for movie in movies_data:
        # print(movie)
       # connection = sqlite3.connect("movies.db")
        cursor = connection.cursor()
         # SQL to insert data
        cursor.execute('''
        INSERT INTO movies (title, year, poster, duration, imdb_rating, genre, stars, director)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?)
         ''', (
        movie['title'],
        movie['year'],
        movie['poster'],
        movie['duration'],
        movie['imdb_rating'],
        movie['genre'],
        movie['stars'],
        movie['director']
    ))
    connection.commit()


#use the function to process and save movies to db
process_save_movies(div_movies)
```

```python

import sqlite3

"""
    Creates movie_actors table in SQLite if it doesn't exist ,
    and insert actor id (from actors table) with movies id (from movies table) to movie_actors table
    creating a many-to-many relationship between actors and movies.
"""

# Connect to the database
connection = sqlite3.connect("movies_db.db")
cursor = connection.cursor()

# Step 1: Create the `movie_actors` table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS movie_actors (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        actor_id INTEGER,
        movie_id INTEGER,
        FOREIGN KEY(actor_id) REFERENCES actors(id),
        FOREIGN KEY(movie_id) REFERENCES movies(id),
        UNIQUE(actor_id, movie_id)
    )
''')
connection.commit()

try:
    # Step 2: Fetch movies and split the stars field
    cursor.execute("SELECT id, stars FROM movies")  # Assuming `movies` table has `stars` and `id` columns
    movies = cursor.fetchall()  # List of (movie_id, stars) tuples

    for movie_id, actors_field in movies:
        # Debug: Print the actors field
        print(f"Processing movie_id {movie_id}, actors_field: {actors_field}")

        # Skip if actors_field is None or empty
        if not actors_field:
            print(f"No actors listed for movie_id {movie_id}. Skipping...")
            continue

        try:
            # Split the actors field into a list
            actors_list = [actor.strip() for actor in actors_field.split(',')]
        except Exception as e:
            print(f"Error processing actors for movie_id {movie_id}: {e}")
            continue

        # Step 3: Match each actor name with their ID in the `actors` table
        for actor_name in actors_list:
            cursor.execute("SELECT id FROM actors WHERE actor_name = ?", (actor_name,))
            actor = cursor.fetchone()  # Fetch the actor's ID

            if actor:
                actor_id = actor[0]

                # Step 4: Insert actor_id and movie_id into the `movie_actors` table
                try:
                    cursor.execute('''
                        INSERT INTO movie_actors (actor_id, movie_id)
                        VALUES (?, ?)
                    ''', (actor_id, movie_id))
                except sqlite3.IntegrityError:
                    print(f"Duplicate entry for actor_id {actor_id} and movie_id {movie_id}. Skipping...")
            else:
                print(f"Actor '{actor_name}' not found in the actors table.")

    # Commit the changes
    connection.commit()
    print("Actor-movie relationships successfully added to `movie_actors` table.")

except sqlite3.Error as e:
    print(f"An error occurred: {e}")

finally:
    # Close the database connection
    connection.close()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the database
connection = sqlite3.connect("movies_db.db")
cursor = connection.cursor()

# Query: Number of movies per actor, limited to top 8
query = '''
SELECT
    actors.actor_name AS actor_name,
    COUNT(movie_actors.movie_id) AS movie_count
FROM
    movie_actors
JOIN
    actors ON movie_actors.actor_id = actors.id
GROUP BY
    actors.id
ORDER BY
    movie_count DESC
LIMIT 8;
'''

# Execute the query and load the results into a DataFrame
df = pd.read_sql_query(query, connection)

# Close the connection
connection.close()

# Step 1: Define custom palette and extend it if necessary
custom_palette = ['#B20710', '#E50914', '#2D3047']
colors = custom_palette * (len(df) // len(custom_palette) + 1)  # Repeat palette if needed

# Step 2: Plot the results with custom colors
plt.figure(figsize=(12, 6))
plt.bar(df['actor_name'], df['movie_count'], color=colors[:len(df)])
plt.xlabel('Actor Name', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.title('Top 8 Actors by Number of Movies', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**IMDb Ratings of Movies for Top 8 Actors (by number of movies)**

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the database
connection = sqlite3.connect("movies_db.db")
cursor = connection.cursor()

# Step 1: Get the top 8 actors by the number of movies
top_actors_query = '''
SELECT
    actors.actor_name AS actor_name,
    actors.id AS actor_id,
    COUNT(movie_actors.movie_id) AS movie_count
FROM
    movie_actors
JOIN
    actors ON movie_actors.actor_id = actors.id
GROUP BY
    actors.id
ORDER BY
    movie_count DESC
LIMIT 8;
'''
top_actors_df = pd.read_sql_query(top_actors_query, connection)

# Step 2: Get IMDb ratings for movies associated with the top 8 actors
imdb_query = '''
SELECT
    actors.actor_name AS actor_name,
    movies.title AS movie_title,
    movies.imdb_rating AS imdb_rating
FROM
    movie_actors
JOIN
    actors ON movie_actors.actor_id = actors.id
JOIN
    movies ON movie_actors.movie_id = movies.id
WHERE
    actors.id IN (
        SELECT
            actors.id
        FROM
            movie_actors
        JOIN
            actors ON movie_actors.actor_id = actors.id
        GROUP BY
            actors.id
        ORDER BY
            COUNT(movie_actors.movie_id) DESC
        LIMIT 8
    )
ORDER BY
    actors.actor_name, movies.imdb_rating DESC;
'''
imdb_df = pd.read_sql_query(imdb_query, connection)

# Close the connection
connection.close()

# Step 3: Plot IMDb ratings for movies of the top 8 actors
plt.figure(figsize=(14, 8))
for actor in imdb_df['actor_name'].unique():
    actor_movies = imdb_df[imdb_df['actor_name'] == actor]
    plt.scatter(
        actor_movies['movie_title'],
        actor_movies['imdb_rating'],
        label=actor,
        s=100,
        alpha=0.8
    )

plt.title('IMDb Ratings of Movies for the Top 8 Actors', fontsize=16)
plt.xlabel('Movies', fontsize=12)
plt.ylabel('IMDb Rating', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Actor', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the database
connection = sqlite3.connect("movies_db.db")
cursor = connection.cursor()

#  Get IMDb ratings for movies of Jenna Ortega, Morgan Freeman, and Chris Hemsworth
imdb_query = '''
SELECT
    actors.actor_name AS actor_name,
    movies.title AS movie_title,
    movies.imdb_rating AS imdb_rating
FROM
    movie_actors
JOIN
    actors ON movie_actors.actor_id = actors.id
JOIN
    movies ON movie_actors.movie_id = movies.id
WHERE
    actors.actor_name IN ('Jenna Ortega', 'Morgan Freeman', 'Chris Hemsworth')
ORDER BY
    actors.actor_name, movies.imdb_rating DESC;
'''
imdb_df = pd.read_sql_query(imdb_query, connection)

# Close the connection
connection.close()

# Define custom palette
custom_palette = ['#B20710', '#E50914', '#2D3047']
actor_colors = dict(zip(imdb_df['actor_name'].unique(), custom_palette))

#  Plot IMDb ratings for movies of the three actors
plt.figure(figsize=(14, 8))

# Create bars for each actor's movies with their assigned color
for actor in imdb_df['actor_name'].unique():
    actor_movies = imdb_df[imdb_df['actor_name'] == actor]
    plt.bar(
        actor_movies['movie_title'],
        actor_movies['imdb_rating'],
        label=actor,
        color=actor_colors[actor],
        alpha=0.8
    )

# Add titles, labels, and legend
plt.title('IMDb Ratings of Movies for Jenna Ortega, Morgan Freeman, and Chris Hemsworth', fontsize=16)
plt.xlabel('Movies', fontsize=12)
plt.ylabel('IMDb Rating', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Actor', fontsize=10)
plt.tight_layout()
plt.show()
```

**IMDb Rating Vs Duration**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3  # or another library to connect to your database
# Connect to your database
connection = sqlite3.connect('movies_db.db')  # Replace with your database file
query = "SELECT imdb_rating,  duration FROM movies"

# Execute the query and load the data into a DataFrame
movies_df = pd.read_sql_query(query, connection)


plt.figure(figsize=(10, 6))
sns.scatterplot(data=movies_df, x='duration', y='imdb_rating', color='#2D3047',  label="Movies", edgecolor='black',alpha=0.8, s=50)

# Add a trend line for better insight
sns.regplot(data=movies_df, x='duration', y='imdb_rating', scatter=False, color='#E50914',  ci=None, label="Trendline")

# Add labels and title
plt.title('IMDB Rating vs Duration', fontsize=16)
plt.xlabel('Duration (minutes)', fontsize=12)
plt.ylabel('IMDB Rating', fontsize=12)
plt.grid(alpha=0.3)


# Add correlation coefficient annotation
correlation = movies_df['duration'].corr(movies_df['imdb_rating'])
plt.annotate(
    f"Correlation: {correlation:.2f}",
    xy=(0.7, 0.1),  # Adjust based on layout
    xycoords='axes fraction',
    fontsize=10,
    bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white")
)


plt.tight_layout()

plt.show()
```
